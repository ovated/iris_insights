```python
def fibonacci_sequence(n):
    """Fibonacci sequence up to n-th number"""
    # Ensure n is not less than or equal to 0, or equal to 1
    if n <= 0:
        return []
    elif n == 1:
        return [0]

    sequence = [0,1]
    # Compute the fibonacci sequence
    for i in range(2,n):
        sequence.append(sequence[-1] + sequence[-2])
    return sequence
```

```python
# Test
fibonacci_sequence(10)
```

```python
# Import relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# To read iris dataset from, not for plotting this time
import seaborn as sns
```

```python
# Load the Iris dataset via seaborn
df = sns.load_dataset("iris")
```

```python
# Display the first five rows with head
df.head()
```

```python
# Create a scatter plot
fig, ax = plt.subplots(figsize=(6,4))

# Group rows by species and plot each group seperately
for species_name, species_frame in df.groupby("species"):
    ax.scatter(species_frame["sepal_length"], species_frame["sepal_width"], label=species_name, s=25)

# Add title
ax.set_title("Iris: Sepal Length vs Sepal Width")

# Add axis labels
ax.set_xlabel("Sepal Length (cm)")
ax.set_ylabel("Sepal Width (cm)")

# Add label information
ax.legend(title="Species")

plt.show()
```

### Data Analysis

```python
# Compute the mean and median petal length for each species
petal_length_stats = (
    df.groupby("species")["petal_length"]
    .agg(mean_petal_length="mean", median_petal_length="median")
    .reset_index()
)
petal_length_stats
```

### Use of Magic Commands

```python
def custom_sort(arr):
    """Return a sorted copy of an array"""
    return sorted(arr)

# Convert th epetal_width column to a nunmpy array
petal_width_array = df["petal_width"].to_numpy()

# Time it using the time magic command
%timeit custom_sort(petal_width_array)
```
